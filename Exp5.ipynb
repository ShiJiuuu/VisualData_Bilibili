```python
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_NOTEBOOK
```

```python
from pyecharts.charts import Bar
from pyecharts import options as opts
from pyecharts.charts import Pie

years = ["2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018"]
user_counts = [3095, 4245, 6673, 10701, 13642, 31368, 40949, 41776, 56213, 64143]

bar = (
    Bar(init_opts=opts.InitOpts(theme="ROMANTIC", width="800px", height="400px"))
    .add_xaxis(years)
    .add_yaxis("Register Users", user_counts)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="The distribution of user registration time in the Hupu community"),
        xaxis_opts=opts.AxisOpts(name="Register Time"),
        yaxis_opts=opts.AxisOpts(name="Users Num"),
    )
)

bar.render_notebook()

gender_data = [("Male", 4.5), ("Female", 95.5)]

pie = (
    Pie()
    .add("", gender_data, radius=["40%", "70%"])
    .set_global_opts(
        title_opts=opts.TitleOpts(title="The ratio of male and female users in the Hupu community"),
        legend_opts=opts.LegendOpts(orient="vertical", pos_right="10%", pos_top="middle")
    )
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {d}%"))
)

pie.render_notebook()
```
