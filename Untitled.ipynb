```python
from pyecharts.charts import Pie
from pyecharts import options as opts
from pyecharts.globals import ThemeType

# 模拟数据（请替换为实际数据）
categories = ["NBA", "CBA", "足球", "步行街", "电竞", "影视", "汽车", "数码"]
post_counts = [12500, 6800, 9200, 18500, 7500, 4200, 3100, 2800]

# 创建圆环图
pie = (
    Pie(init_opts=opts.InitOpts(theme=ThemeType.INFOGRAPHIC, width="100%", height="600px"))
    .add(
        series_name="发帖数量",
        data_pair=[list(z) for z in zip(categories, post_counts)],
        radius=["40%", "70%"],  # 内半径和外半径控制圆环宽度
        label_opts=opts.LabelOpts(
            formatter="{b}: {c} ({d}%)",  # 显示板块名、数值和百分比
            font_size=12
        )
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title="虎扑各板块发帖数分布",
            subtitle="近3个月抽样数据",
            pos_left="center"
        ),
        legend_opts=opts.LegendOpts(
            orient="vertical",
            pos_top="15%",
            pos_left="2%"
        ),
        tooltip_opts=opts.TooltipOpts(
            trigger="item",
            formatter="{a}<br/>{b}: {c} ({d}%)"
        )
    )
    .set_series_opts(
        itemstyle_opts={
            "borderRadius": 8,  # 圆角效果
            "borderColor": "#fff",
            "borderWidth": 2
        }
    )
)

# 在Jupyter中显示
pie.render_notebook()
map_chart.render("hupu_map.html")
```
