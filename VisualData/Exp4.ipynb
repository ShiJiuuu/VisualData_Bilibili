```python
import matplotlib.pyplot as plt

days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
prices = [44.98, 45.02, 44.32, 41.05, 42.08]

fig = plt.figure()
ax = fig.add_axes([0.2, 0.2, 0.5, 0.5])

ax.plot(days, prices, marker='o')

ax.set_xticks(days)
ax.set_xticklabels(days)

ax.tick_params(axis='both', direction='in', width=2)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.set_ylabel("Closing")
ax.set_title("A Line Chart of a Stock's Weekly Closing Price")

plt.show()
```
