### 面向函数的方式

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.figure(figsize=(8, 4))
plt.plot(x, y_sin, label='Sin', color='blue')
plt.plot(x, y_cos, label='Cos', color='red')
plt.title('Sin and Cos Func Graph')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 面向对象的方式

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x, y_sin, label='Sin', color='blue')
ax.plot(x, y_cos, label='Cos', color='red')
ax.set_title('Sin and Cos Func Graph')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.grid(True)

fig.tight_layout()
plt.show()
```
