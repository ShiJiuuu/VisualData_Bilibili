```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
data_2018 = [39, 20, 40, 38, 42, 43, 41, 41, 45, 48, 52, 50]
data_2019 = [45, 28, 48, 49, 50, 51, 50, 50, 51, 52, 70, 65]

plt.style.use('fivethirtyeight')

plt.figure(figsize=(10, 6))
plt.plot(months, data_2018, color="#8B0000", marker='^', linestyle='--', linewidth=1.5, label='2018')
plt.plot(months, data_2019, color="#006374", marker='v', linestyle='-', linewidth=1.5, label='2019')

plt.xlabel("Month")
plt.ylabel("Business volume (100M Pieces)")
plt.title("Trend Chart of Express Business Volume in 2018 and 2019")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
